```python
import matplotlib.pyplot as plt

size = [2441,2312,1031,1233]  
label = ['A형','B형','C형','D형']   #반시계 방향 (파이 차트의 시작점은 3시 방향임)


plt.rc('font', family='Malgun Gothic') #그래프에 한글 표시

plt.axis = ('equal')                   #좀더 동그란, 균일한 원 그리기
plt.pie(size , labels = label)       #(파이) '각 값이 차지 하는 비율로 '원 그리기

plt.show()
```

```python
import matplotlib.pyplot as plt

size = [2441,2312,1031,1233]  
label = ['A형','B형','C형','D형']   #반시계 방향 (파이 차트의 시작점은 3시 방향임)


plt.rc('font', family='Malgun Gothic') #그래프에 한글 표시

plt.axis = ('equal')                   #좀더 동그란, 균일한 원 그리기

#(파이) 
#autopct : 자동 비율 표시
#explod  : 돌출 표시
plt.pie(size , labels = label, autopct='%.1f%%' , explode=[0,0,0.1,0])   

#범례 추가
plt.legend()  

plt.show()
```

```python
import matplotlib.pyplot as plt

size = [2441,2312,1031,1233]  
label = ['A형','B형','C형','D형']   #반시계 방향 (파이 차트의 시작점은 3시 방향임)
color = ['crimson','darkcyan']

plt.rc('font', family='Malgun Gothic') #그래프에 한글 표시

#색상 지정



plt.axis = ('equal')                   #좀더 동그란, 균일한 원 그리기

#(파이) 
#autopct : 자동 비율 표시
#explod  : 돌출 표시
#color   : 색상 지정
#startangle : 시작점  (최소는 3시 방향)    , startangle = 90 는 시계 반대 방향으로 90도. 즉 12시가 됨
plt.pie(size , labels = label, autopct='%.1f%%' , explode=[0,0,0.1,0], colors= color , startangle = 90)   

#범례 추가
plt.legend()  

plt.show()
```
