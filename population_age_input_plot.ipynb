```python
import csv
import matplotlib.pyplot as plt

f = open('D:/99. STUDY/02. people age analisys/202008_202008_age_month.csv','r', encoding='cp949')
data = csv.reader(f)

result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해 주세요:')  #문자열 입력 받기

for row in data:
    if name in row[0]:  #해당 문자열 포함 여부
        print(row)
        for i in row[3:]:
            result.append(int(i.replace(',','')))

plt.style.use('ggplot')                # plot 의 격자 문의
plt.rc('font',family='Malgun Gothic')  #한글 제목 표기, 폰트                 
plt.title(name + '지역의 인구 구조')
plt.plot(result)
plt.show()
```
